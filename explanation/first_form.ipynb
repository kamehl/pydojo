##ex8 æ¸©åº¦å¤‰æ›

è¯æ°ã¨æ‘‚æ°ã‚’ç›¸äº’ã«å¤‰æ›ã™ã‚‹é–¢æ•°ã‚’ä½œæˆã›ã‚ˆ

ãŸã ã—ã€è¯æ°ã¯ã€ŒFã€ã‚’ã€æ‘‚æ°ã¯ã€ŒCã€ã‚’ã€ãã‚Œãã‚Œæ•°å­—ã®å¾Œã«ä»˜ã‘ãŸæ–‡å­—åˆ—ã§è¡¨ã™ã“ã¨ã¨ã™ã‚‹

ãªãŠã€è¯æ°ã¨æ‘‚æ°ã®å¤‰æ›å¼ã¯æ¬¡ã®ã¨ãŠã‚Š

>$(F - 32) Ã— 5 Ã· 9 = C$

å¤‰æ›å¾Œã®æ•°å€¤ã¯å°æ•°ç‚¹1æ¡ã¾ã§ï¼ˆæœªæº€ã¯å››æ¨äº”å…¥ï¼‰

â€»ã‚‚ã—å¯èƒ½ãªã‚‰ã€å¤‰æ›ã§ããªã„ä¸æ˜ãªå…¥åŠ›ã«ã¯ã€Œç„¡åŠ¹ã€ã‚’è¿”ã™ã“ã¨

````python
# å››æ¨äº”å…¥ãªã‚‰ãƒ•ã‚©ãƒ¼ãƒãƒƒãƒˆã§ã‚‚å¯èƒ½
def func8(s):
    try:
        if s[-1] in "fF":
            t = int(s[:-1])
            x = (t - 32) * 5 / 9
            return f"{x:.1f}C"
        elif s[-1] in "cC":
            t = int(s[:-1])
            x = 9 * t / 5 + 32
            return f"{x:.1f}F"
        else:
            raise
    except:
        return "ç„¡åŠ¹"
````
